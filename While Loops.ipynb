### While Loops

Line of code executes over and over again while the condition is True. Compared to if statements: if statements checks if its True, runs code and then ends. While loops iterate the code over and over while condition is True and only then ends.

Can have inputs in the while loop block.

```python
spam = 0
while spam < 5:
    print ('Hello World')
    spam = spam + 1
```

Iteration: each repetition. In this example, the code does 5 iterations before it no longer satisfies the condition and ends.

```python
name = ''
while name != 'your name':
    print('Please type your name.')
    name = input()
print('Thank you!')
```

When the input is 'your name' the while condition is evaluated to False, because name == 'your name' and then the loop doesn't run and we get the last print statement which is out of the loop.

```python
while True:
    print('Hello!')
```

Prints for ever, need to stop the code using Ctrl+C on terminal or the stop button on jupyter notebook.

### Break Statements

Causes execution to jump straight out of loop - avoids infinite exectitions like the one above. Doesn't even check condition again, just jumps straight to line right after the loop.

```python
name = ''
while True:
    print('Please type your name.')
    name = input()
    if name == 'your name':
        break
print('Thank you!')
    
    
```

The condition will never be False, because True always evaluates to True. So the code wont ever break in that way. Instead, the if statement checks that if name == 'your name' then it executes break statement. Execution jumps out of loop and prints our Thank you. If the user doesn't write 'your name' then the code prints out please type your name again because the condition is always true. It would do this indefinetely without the if statement.

### Continue Statements

Inside an if statement, if the condition is satisfied, then the program executes the continue statement, which takes execution back to while statement without executing what is after it.

In this case, the execution will no run to the last print statement for spam == 3. When this happens, code just runs back to while loop for spam == 4.

```python
spam = 0
while spam < 5:
    spam = spam + 1
    if spam == 3:
        continue
    print ('spam is ' + str(spam))
```
